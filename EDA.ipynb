Numeric-Numeric Data

Scatter Plot

```python
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip'
)
```

```python
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',
    style='smoker'
)
```

Numeric-Categoric Data

BarPlot


```python
titanic.head()

```

```python
sns.barplot(
    data=titanic,
    x='sex',
    y='survived'
)
```

```python
sns.barplot(
    data=titanic,
    x='pclass',
    y='age',
    hue='survived'
)
```

Numeric-Categoric Data

```python
sns.boxplot(
    data=tips,
    x='day',
    y='total_bill'
)
```

```python
sns.boxplot(
    data=titanic,
    x='sex',
    y='age',
    hue="survived"
)
```

```python
sns.heatmap(pd.crosstab(titanic['sex'],titanic['survived']))
```

```python
sns.heatmap(pd.crosstab(titanic['pclass'],titanic['survived']))
```
